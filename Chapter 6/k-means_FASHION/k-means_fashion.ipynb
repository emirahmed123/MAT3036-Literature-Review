```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
from sklearn.manifold import TSNE
from sklearn.metrics import normalized_mutual_info_score
from sklearn.metrics import adjusted_rand_score
```

Training Samples

```python
# Transforms
transform = transforms.ToTensor()

fashion = datasets.FashionMNIST(root="./data", train=True, download=True, transform=transform)

loader = DataLoader(fashion, batch_size=len(fashion), shuffle=False)
```

```python
images, labels = next(iter(loader))
```

```python
print(images.shape)
```

```python
X = images.view(images.shape[0], -1).numpy()
```

```python
X.shape
```

```python
y = labels.numpy()
```

```python
y.shape
```

```python
kmeans = KMeans(n_clusters=10, random_state=0)
cluster_ids = kmeans.fit_predict(X)
```

```python
nmi_score = normalized_mutual_info_score(y, cluster_ids)
print(f"NMI: {nmi_score:.4f}")
```

```python
ari = adjusted_rand_score(y, cluster_ids)
print(f"ARI: {ari:.4f}")
```
