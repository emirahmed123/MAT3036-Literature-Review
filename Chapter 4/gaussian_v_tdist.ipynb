```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
```

```python
# Generate x values
x = np.linspace(-10, 10, 500)
```

```python
# Gaussian (mean=0, std=1)
gaussian = norm.pdf(x, loc=0, scale=1)

# Student t-distribution (df=1)
t_dist = t.pdf(x, df=1)
```

```python
# Plot
plt.figure(figsize=(10, 5))

plt.plot(x, gaussian, label="Gaussian (mean=0, std=1)")
plt.plot(x, t_dist, label="t-distribution (df=1)")

# Labels and title
plt.xlabel("x", fontsize=22)
plt.ylabel("Density", fontsize=22)
plt.title("Gaussian vs Student t-Distribution", fontsize=25, fontweight="bold")

# Tick size
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

# Legend
plt.legend(fontsize=16)

plt.tight_layout()

# Save figure
# plt.savefig("gaussian_vs_tdist.pdf", bbox_inches='tight')

plt.show()
```
